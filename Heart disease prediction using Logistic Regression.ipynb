# Heart disease prediction using Logistic Regression

### **GOAL: To create a Classification Model that can predict whether or not a person has presence of heart disease based on physical features of that person (age,sex, cholesterol, etc...)**

## Imports


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data

This database contains 14 physical attributes based on physical testing of a patient. Blood samples are taken and the patient also conducts a brief exercise test. The "goal" field refers to the presence of heart disease in the patient. It is integer (0 for no presence, 1 for presence). In general, to confirm 100% if a patient has heart disease can be quite an invasive process, so if we can create a model that accurately predicts the likelihood of heart disease, we can help avoid expensive and invasive procedures.

Content

Attribute Information:

* age
* sex
* chest pain type (4 values)
* resting blood pressure
* serum cholestoral in mg/dl
* fasting blood sugar > 120 mg/dl
* resting electrocardiographic results (values 0,1,2)
* maximum heart rate achieved
* exercise induced angina
* oldpeak = ST depression induced by exercise relative to rest
* the slope of the peak exercise ST segment
* number of major vessels (0-3) colored by flourosopy
* thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
* target:0 for no presence of heart disease, 1 for presence of heart disease

Original Source: https://archive.ics.uci.edu/ml/datasets/Heart+Disease

## Train | Test Split and Scaling

**Separating the features from the labels into 2 objects, X and y.**

```python
X = df.drop('target',axis=1)
y = df['target']
```

**Performing a train test split on the data, with the test size of 10% and a random_state of 101.**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

**Creating a StandardScaler object and normalize the X train and test set feature data.**

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Logistic Regression Model

**Creating a Logistic Regression model and using Cross-Validation to find a well-performing C value for the hyper-parameter search.**

```python
from sklearn.linear_model import LogisticRegressionCV 
```

```python
# help(LogisticRegressionCV)
```

```python
log_model = LogisticRegressionCV()
```

```python
log_model.fit(scaled_X_train,y_train)
```

**Reporting back the search's optimal parameters, specifically the C value.** 

```python
log_model.C_
```

```python
log_model.get_params()
```

### Coeffecients

**Reporting back the model's coefficients.**

```python
log_model.coef_
```

**Creating a visualization of the coefficients by using a barplot of their values.**

```python
coefs = pd.Series(index=X.columns,data=log_model.coef_[0])
```

```python
coefs = coefs.sort_values()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values);
```

---------

## Model Performance Evaluation

**Let's now evaluate the model on the remaining 10% of the data, the test set.**

**Creating the following evaluations:**
* Confusion Matrix Array
* Confusion Matrix Plot
* Classification Report

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
plot_confusion_matrix(log_model,scaled_X_test,y_test)
;
```

```python
print(classification_report(y_test,y_pred))
```

### Performance Curves

**Creating both the precision recall curve and the ROC Curve.**

```python
from sklearn.metrics import plot_precision_recall_curve,plot_roc_curve
```

```python
plot_precision_recall_curve(log_model,scaled_X_test,y_test)
;
```

```python
plot_roc_curve(log_model,scaled_X_test,y_test)
;
```
